1. ID_Contrato

O que é: Identificador único da operação.

Utilidade no Risco: Serve para rastrear a evolução do crédito ao longo do tempo (safras) e garantir que não haja duplicidade no cálculo do RWA.

2. Produto

O que é: Categoria do crédito (Emprestimo_Pessoal, Cartao_Credito, Parcelado_Loja).

Utilidade no Risco: Diferentes produtos possuem diferentes RWs (Risk Weights) para o capital regulatório e comportamentos de inadimplência distintos.

3. Valor_Nocional

O que é: O saldo devedor atual (principal + juros acruados).

Utilidade no Risco: É a base para o cálculo da EAD (Exposure at Default). No cálculo do EVE, este é o valor que será projetado e trazido a valor presente.

4. Taxa_Contratada_AA

O que é: Taxa de juros anual (Anualizada) acordada com o cliente no momento da assinatura.

Utilidade no Risco: Define a geração de receita (NII). Se a taxa contratada for fixa (PRE) e o custo de captação subir, essa coluna revela o risco de compressão da margem.

5. Prazo_Remanescente_Meses

O que é: Quantos meses faltam para o contrato encerrar.

Utilidade no Risco: Crucial para o cálculo do EVE e Duration. Quanto maior o prazo, mais sensível o valor desse contrato é a mudanças na taxa de juros (maior risco de mercado).

6. Indexador

O que é: Define se o contrato é PRE (taxa fixa) ou POS-CDI (taxa que oscila com a Selic).

Utilidade no Risco: Essencial para o Gap de Reprecificação. Contratos PRE geram risco de descasamento em cenários de alta de juros; contratos POS protegem o NII, mas podem aumentar a inadimplência se o cliente não aguentar a subida da parcela.

7. Rating_2682

O que é: Nota de crédito baseada na Resolução CMN 2.682 (AA a H).

Utilidade no Risco: Define o percentual de Provisão que o banco deve manter. Um Rating 'H' exige 100% de reserva, enquanto um 'AA' exige 0%.

8. Dias_Atraso

O que é: Há quantos dias o cliente não paga a parcela vencida.

Utilidade no Risco: É o principal driver para a mudança automática de Rating e impacto imediato no P&L através das perdas de crédito (provisões).
